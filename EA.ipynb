## GENDER ANALYSIS

```python
avg_age = df.groupby('Gender')['Age'].mean().reset_index(name='avg')

print(avg_age)
```

```python
from matplotlib.gridspec import GridSpec


# Set the size of the plots
plt.figure(figsize=(8, 12))

# Create a GridSpec to organize the layout
gs = GridSpec(2, 1, height_ratios=[1, 1])

# Boxplot with attrition on the X-axis and Job Satisfaction on the Y-axis
ax1 = plt.subplot(gs[0])
sns.boxplot(data=df, x='Attrition', y='JobSatisfaction', hue='Attrition', palette=['#FA5858', '#9FF781'], ax=ax1)
ax1.set_title('Boxplot of Job Satisfaction by Attrition and Gender')
ax1.legend_.remove()
sns.despine(ax=ax1)
ax1.set_xlabel('Attrition')
ax1.set_ylabel('Job Satisfaction')

# Facet by Gender
sns.boxplot(data=df, x='Attrition', y='JobSatisfaction', hue='Attrition', palette=['#FA5858', '#9FF781'], ax=ax1)
ax1.set_title('Boxplot of Job Satisfaction by Attrition and Gender')
ax1.legend_.remove()
sns.despine(ax=ax1)
ax1.set_xlabel('Attrition')
ax1.set_ylabel('Job Satisfaction')

# Distribution of Job Satisfaction
ax2 = plt.subplot(gs[1])
sns.kdeplot(data=df, x='JobSatisfaction', color="#013ADF", fill=True, ax=ax2)
ax2.set_xlim(1, 4)
ax2.set_title('Distribution of Job Satisfaction')
sns.despine(ax=ax2)
ax2.set_xlabel('Job Satisfaction')
ax2.set_ylabel('Density')

plt.tight_layout()
plt.show()
```
